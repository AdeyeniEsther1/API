## API
Stands for Application programming interface: it allows different software that are built differently to communicate with one another.
- Work in a server client mode
- Json is the end product of the request
- REST: This is the means of communication, done through the web
- GET: used to request data
- POST: used to write data to the server(new data)
- DELETE: delete data
- PUT: write(update data)

https://api.stackexchange.com/

```python
import webbrowser
import urllib.request
```

```python
##Define APOD
apodurl = "https://api.nasa.gov/planetary/apod?"
mykey = "api_key=xCGILT6hKAanRWkuMZPbjZuSIqTwFmitIhNIS29N"
```

```python
# call the webservice
apodurlobj = urllib.request.urlopen(apodurl + mykey)
```

```python
## read the file-like object
apodread = apodurlobj.read()
```

```python
apodread
```

```python
# decode json to python data structure
decodeapod = json.loads(apodread.decode("utf-8"))
```

```python
# display the python data
print("\n\nConverted python data")
print(decodeapod)
```
